```python
# Türev(Derivative) Nedir ?
# Türev, bir fonksiyonun belirli noktadaki eğimini(slope) veya değişim hızını ölçer.
# yani bir noktadaki türev, fonksiyonun o noktada ne kadar hızlı değiştiğini gösterir.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Örnek Fonksiyon
def f(x):
    return (3*x**2 - 4*x+ 5)

print(f"x = {3} için fonksiyonun değeri: {f(3)}")   # Tek değer

#  -5 ile 5 arasındaki değerlerin hepsini f() uygulandı ve değerlerin grafiği çizildi
x = np.arange(-5, 5, 0.25)   
y = f(x)
print(y)
plt.plot(x,y)
```

```python
# Şimdi ise önceki fonksiyonun türevini bulalım
# Türev formülü = f'(x) = lim h→0 (f(x + h) - f(x))/h  
# 1. f(x + h) - f(x) = Fonksiyonun iki değer arasında ne kadar değiştini ifade eder.yani iki değer arasındaki farkı hesaplar
# 2. (f(x + h) - f(x))/h = Fonksiyonun iki değer arasındaki değişim oranını(eğimini) hesaplar.
# Bu, fonksiyonun 𝑥 noktasındaki eğimini belirlemek için kullanılır.
# 3. lim h→0 = Türev hesaplamalarında,  ℎ'nin sıfıra yaklaşan küçük bir "değişim miktarını" temsil eder. 
# Bu limit, fonksiyonun tam olarak 𝑥 noktasındaki eğimini bulmamıza olanak tanır.

def f(x):
    return (3*x**2 - 4*x+ 5)
    
h = 0.0001
x = 3
derivative = (f(x + h) - f(x)) / h
print(f"{f(x)}'nin Türevi: {derivative}")
```

```python
# çoklu değerler(noktalar) için türev

def f(x):
    return (3*x**2 - 4*x + 5)
    
h = 1e-4           # 0.0001
X = np.arange(-3, 4, 1)
derivative = (f(X + h) - f(X)) / h
for x, d in zip(X, derivative):
    print(f"{f(x)} Değerinin Türevi: {d}")

plt.plot(X, f(X), label="f(x) = 3x² - 4x + 5", color="blue")       # Fonksiyon grafiği
plt.plot(X, derivative, label="Türev", color='orange', linestyle='--')  # Türev grafiği
plt.title("Fonksiyon ve Türev Grafiği")
plt.xlabel("X (Girdi)") 
plt.ylabel("Fonksiyon Değeri ve Türev Değeri")
plt.axhline(0, color='black', linewidth=0.75, ls='--')
plt.axvline(0, color='black', linewidth=0.75, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
# Birden fazla değişken için türev değerleri hesaplama
# her bir girdi için türev ayrı ayrı hesaplanıyor
# Fonksiyonun bir girdiye göre türevi diğer girdiler sabit tutulurken yapılır.

def f(a, b, c):
    return a*b + c

h = 1e-4

# Girdi değerleri
a = 2
b = -3
c = 10

# d'nin başlangıç değerleri
d = f(a, b, c)

# a'ya göre türev
dA = f(a + h, b, c)
slope_a = (dA - d) / h
print(f"A'ya Göre Türevi: {slope_a}")       # d/dA

# b'ye göre türev
dB = f(a, b + h, c)
slope_b = (dB - d) / h
print(f"B'ye Göre Türevi: {slope_b}")       # d/dB

# c'ye göre türev
dC = f(a, b, c + h)
slope_c = (dC - d) / h
print(f"C'ye Göre Türevi: {slope_c}")       # d/dC
```
