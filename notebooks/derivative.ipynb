```python
# T√ºrev(Derivative) Nedir ?
# T√ºrev, bir fonksiyonun belirli noktadaki eƒüimini(slope) veya deƒüi≈üim hƒ±zƒ±nƒ± √∂l√ßer.
# yani bir noktadaki t√ºrev, fonksiyonun o noktada ne kadar hƒ±zlƒ± deƒüi≈ütiƒüini g√∂sterir.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# √ñrnek Fonksiyon
def f(x):
    return (3*x**2 - 4*x+ 5)

print(f"x = {3} i√ßin fonksiyonun deƒüeri: {f(3)}")   # Tek deƒüer

#  -5 ile 5 arasƒ±ndaki deƒüerlerin hepsini f() uygulandƒ± ve deƒüerlerin grafiƒüi √ßizildi
x = np.arange(-5, 5, 0.25)   
y = f(x)
print(y)
plt.plot(x,y)
```

```python
# ≈ûimdi ise √∂nceki fonksiyonun t√ºrevini bulalƒ±m
# T√ºrev form√ºl√º = f'(x) = lim h‚Üí0 (f(x + h) - f(x))/h  
# 1. f(x + h) - f(x) = Fonksiyonun iki deƒüer arasƒ±nda ne kadar deƒüi≈ütini ifade eder.yani iki deƒüer arasƒ±ndaki farkƒ± hesaplar
# 2. (f(x + h) - f(x))/h = Fonksiyonun iki deƒüer arasƒ±ndaki deƒüi≈üim oranƒ±nƒ±(eƒüimini) hesaplar.
# Bu, fonksiyonun ùë• noktasƒ±ndaki eƒüimini belirlemek i√ßin kullanƒ±lƒ±r.
# 3. lim h‚Üí0 = T√ºrev hesaplamalarƒ±nda,  ‚Ñé'nin sƒ±fƒ±ra yakla≈üan k√º√ß√ºk bir "deƒüi≈üim miktarƒ±nƒ±" temsil eder. 
# Bu limit, fonksiyonun tam olarak ùë• noktasƒ±ndaki eƒüimini bulmamƒ±za olanak tanƒ±r.

def f(x):
    return (3*x**2 - 4*x+ 5)
    
h = 0.0001
x = 3
derivative = (f(x + h) - f(x)) / h
print(f"{f(x)}'nin T√ºrevi: {derivative}")
```

```python
# √ßoklu deƒüerler(noktalar) i√ßin t√ºrev

def f(x):
    return (3*x**2 - 4*x + 5)
    
h = 1e-4           # 0.0001
X = np.arange(-3, 4, 1)
derivative = (f(X + h) - f(X)) / h
for x, d in zip(X, derivative):
    print(f"{f(x)} Deƒüerinin T√ºrevi: {d}")

plt.plot(X, f(X), label="f(x) = 3x¬≤ - 4x + 5", color="blue")       # Fonksiyon grafiƒüi
plt.plot(X, derivative, label="T√ºrev", color='orange', linestyle='--')  # T√ºrev grafiƒüi
plt.title("Fonksiyon ve T√ºrev Grafiƒüi")
plt.xlabel("X (Girdi)") 
plt.ylabel("Fonksiyon Deƒüeri ve T√ºrev Deƒüeri")
plt.axhline(0, color='black', linewidth=0.75, ls='--')
plt.axvline(0, color='black', linewidth=0.75, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
# Birden fazla deƒüi≈üken i√ßin t√ºrev deƒüerleri hesaplama
# her bir girdi i√ßin t√ºrev ayrƒ± ayrƒ± hesaplanƒ±yor
# Fonksiyonun bir girdiye g√∂re t√ºrevi diƒüer girdiler sabit tutulurken yapƒ±lƒ±r.

def f(a, b, c):
    return a*b + c

h = 1e-4

# Girdi deƒüerleri
a = 2
b = -3
c = 10

# d'nin ba≈ülangƒ±√ß deƒüerleri
d = f(a, b, c)

# a'ya g√∂re t√ºrev
dA = f(a + h, b, c)
slope_a = (dA - d) / h
print(f"A'ya G√∂re T√ºrevi: {slope_a}")       # d/dA

# b'ye g√∂re t√ºrev
dB = f(a, b + h, c)
slope_b = (dB - d) / h
print(f"B'ye G√∂re T√ºrevi: {slope_b}")       # d/dB

# c'ye g√∂re t√ºrev
dC = f(a, b, c + h)
slope_c = (dC - d) / h
print(f"C'ye G√∂re T√ºrevi: {slope_c}")       # d/dC
```
